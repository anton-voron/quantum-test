```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../train.csv')

display(data.shape)
data.head()
```

```python
print("\nDescriptive statistics:")
summary_statistics = data.describe()

# Calculate basic statistics summary
basic_stats = summary_statistics.T[['count', 'mean', '50%', 'std', 'min', 'max', '25%', '50%', '75%']]
basic_stats.columns = ['Count', 'Mean', 'Median', 'Standard Deviation', 'Min', 'Max', '25%', '50%', '75%']
basic_stats

# Feature distribution
    # - The mean values of the features are approximately in the range of 7 to 250.
    # - The std vary to match among features, means different scales and distributions.
    # - Min for most features are around 0, while max values vary (15 - 499).
    # - The quartiles (25%, 50%, 75%) show the spread of the data within each feature.


# Target Column:
    # - The mean target value is 50.
    # - The std is around 28.9, means a wide spread in the target values.
    # - The min target value is approximately 0.0026, while the max is about 99.9995.
    # - The quartiles indicate that the target values are distributed uniformly between the minimum and maximum values.
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['target'], kde=True, bins=30)
plt.title('Distribution of Target Variable')
plt.show()
```

```python
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Features and Target')
plt.show()


# Features has verly low correlation with the target variable, and among themselves (almost).
# Feature 6 has correlation with feature 8 cor = 0.940016. Actions:
    # Drop feature 6 or 8
    # Conmbine feature 6 and 8 as new feature
    #  Reduce dim with PCA
```

```python
# Extract correlation with the target variable
target_correlation = correlation_matrix['target'].sort_values(ascending=False)

target_correlation_df = target_correlation.to_frame().reset_index()
target_correlation_df.columns = ['Feature', 'Correlation with Target']


target_correlation_df

# Correlation Results:
    # All features have low correlation values with the target variable, both positive and negative.
    # The highest positive correlation is 0.012 (Feature 7), and the highest negative correlation is -0.007 (Feature 40).
# Summary:
    # Low correlation indicates that the features have weak linear relationships with the target variable.
    # Linial model may not capture the relationships well, and more complex models or feature engineering might be needed.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the box plots for a sample of features
sample_columns = data.columns
plt.figure(figsize=(20, 10))
data[sample_columns].boxplot()
plt.title('Box Plot for Sample Features')
plt.xticks(rotation=45)
plt.show()

```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.01, random_state=42)

outliers = iso_forest.fit_predict(data)

outliers = outliers == -1

data_no_outliers_iso = data[~outliers]

num_outliers_iso = outliers.sum()
print(f"Number of outliers detected by Isolation Forest: {num_outliers_iso}")
```

```python
outliers_data = data[outliers]
print(outliers_data.shape)
display(outliers_data.head(10))

outliers_summary = outliers_data.describe()
outliers_summary

# The mean target value for the outliers is 56.49, slightly higher than the mean target value in the full dataset (50.03).
# The std of the target for outliers is higher (32.83 vs. 28.90), indicating more variability among the target values for outliers.
```
